## Apply()

```python
google = pd.read_csv("pandas/google_stock_price.csv", squeeze=True)
google.head()
```

```python
def perform(price):
    if price < 300:
        return "OK"
    elif price >= 300 and price < 700:
        return "Satisfactory"
    else:
        return "Best"
```

```python
google.apply(perform).tail(40)
```

```python
google[2977]
```
